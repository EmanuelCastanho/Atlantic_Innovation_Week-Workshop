### O que são detritos marinhos? [*What is Marine Debris?*] 

**[PT]** Detritos marinhos referem-se a qualquer material sólido persistente que é descartado ou abandonado no ambiente marinho por processos naturais (como desastres naturais) e atividades humanas [1]. Alguns dos tipos mais comuns e prejudiciais de detritos marinhos incluem plásticos que variam muito em tamanho, como sacos, garrafas, equipamento de pesca e partículas de microplástico.

**[EN]** Marine debris refers to any persistent solid material that is disposed or abandoned in the marine environment by natural processes (such as natural disasters) and human activities [1]. Some of the most common and harmful types of marine debris include plastics ranging greatly in size, such as bags, bottles, fishing gear, and microplastic particles.

**Macroplásticos [*Macroplastics*]**

**[PT]** No contexto marinho, milhões de toneladas de lixo plástico entram em nossos oceanos anualmente como micro- a macroplásticos, como resultado, o plástico representa mais de 80 a 90% de todo o lixo encontrado neste ambiente.

Os macroplásticos descrevem grandes items de plástico com diâmetro ≥ 5 mm. Em muitos casos, esses items podem persistir por décadas no oceano, especialmente quando protegidos da radiação UV sob a água ou em sedimentos, representando uma grande parte da massa global de detritos flutuantes [2], [3].

**[EN]** Within the marine context, millions of tonnes of plastic enter our oceans annually as micro- to macroplastic litter, as a result, plastic now accounts for more than 80–90% of all litter found in this environment.

Macroplastics describe large plastic items with a diameter ≥ 5 mm. In many instances, these items can persist for decades in the ocean especially when shielded from UV radiation under water or in sediments, accounting for a large portion of the global mass of floating debris [2], [3].

### Métodos de Detecção e Classificação [*Detection and Classification Methods*]

**[PT]** Um sistema de detecção de detritos marinhos baseado em satélite precisa filtrar todos os pixels que não são água (podem ser usadas máscaras de terra e nuvens), localizar pixels contendo material flutuante e discriminar com sucesso entre detritos antropogénicos (por exemplo, plásticos) e fenómenos naturais (por exemplo, vegetação flutuante). Isso pode ser alcançado usando uma variedade de técnicas, desde índices (por exemplo, FAI e FDI) até *linear unmixing* usando assinaturas espectrais, e *machine learning*. Os sensores de satélites são a principal ferramenta de observação e precisam ser selecionados em termos de tempo de revisita, resolução espacial, disponibilidade espectral e qualidade radiométrica [3].

Quando detritos flutuantes e vegetação são agregados em manchas suficientemente grandes no ambiente marinho, a
detecção através de Sentinel-2 é possível. O instrumento multiespectral (MSI) de 13 bandas a bordo dos satélites Sentinel-2 fornecem tempos de revisita de 2 a 5 dias e resolução espacial de até 10m, permitindo que os macroplásticos sejam distinguíveis de fontes naturais de detritos e da própria água do mar. Esta distinção também é possível devido às propriedades espectrais dos materiais como podemos ver na Fig.4 [2].

**[EN]** A satellite-based marine debris detection system needs to screen out all non-water pixels (land and cloud masks may be used), locate pixels containing floating material and successfully discriminate between anthropogenic debris (e.g. plastics) and natural phenomena (e.g. floating vegetation). This can be achieved using a variety of techniques, from band indices (e.g. FAI and FDI) to linear unmixing using spectral signatures, and machine learning. Satellite remote sensors are the primary observation tool and need to be selected in terms of revisit time, spatial resolution, spectral availability and radiometric quality [3].

When floating debris and vegetation are aggregated into sufficiently large patches in the marine environment,
detection from Sentinel-2 is possible. The 13-band multispectral instrument (MSI) aboard the Sentinel-2 satellites provide revisit times of 2 to 5 days and spatial resolution of up to 10m allowing macroplastics to be distinguishable from natural sources of debris, and seawater itself. This distinction is also possible due to the spectral properties of the materials as we can see in Fig.4 [2].

**Floating Algae Index (FAI)**

**[PT]** A água absorve fortemente a luz nos comprimentos de onda vermelho, próximo do infravermelho (NIR) e infravermelho de ondas curtas (SWIR), enquanto a vegetação flutuando na superfície da água tem maior refletância no NIR do que em outros comprimentos de onda. Estas características permitem distinguir facilmente as algas das águas claras circundantes. As equações 1 e 2 traduzem isso e são usadas para calcular o FAI, um índice considerado uma versão melhorada do NDVI (Normalized Difference Vegetation Index) [4].

**[EN]** Water strongly absorbs light in the red, near infrared (NIR) and shortwave infrared (SWIR) wavelengths, while vegetation floating on the water surface have higher reflectance in the NIR than in other wavelengths. These characteristics allow to easily distinguish algae from surrounding clear water. Equations 1 and 2 translate this and are used to calculate FAI, an index considered to be an improved version of the NDVI (Normalized Difference Vegetation Index) [4].

**Floating Debris Index (FDI)**

**[PT]** Na resolução espacial mais alta do Sentinel-2 MSI, items individuais de detritos provavelmente estarão abaixo dos limites detectáveis até serem agregados em grandes manchas. Para melhorar a detecção de manchas flutuando no oceano nas imagens do Sentinel-2, o FDI foi desenvolvido usando quatro bandas MSI. O FDI é uma versão modificada do FAI onde, no lugar da banda vermelha, onde a clorofila é mais absorvente, é utilizada a banda Red Edge 2 (ver equações 3 e 4). O FDI, portanto, aproveita a diferença entre o NIR e a refletância da linha de base do NIR, apresentando um bom potencial na detecção de plástico offshore [2].

**[EN]** At the highest spatial resolution of the Sentinel-2 MSI, individual items of debris are likely to be below detectable limits until aggregated into large patches. To enhance detection of patches floating on the ocean in Sentinel-2 imagery, the FDI was developed using four MSI bands. FDI is a modified version of the FAI where, in place of the red band, where chlorophyll is most absorptive, the Red Edge 2 band is used (see equations 3 and 4). The FDI thus leverages the difference between the NIR, and the baseline reflectance of NIR, presenting good potential in offshore plastic detection [2].
