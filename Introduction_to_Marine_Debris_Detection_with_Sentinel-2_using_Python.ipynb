## Introduction

### What is Marine Debris?

Marine debris refers to any persistent solid material that is disposed or abandoned in the marine environment by natural processes (such as natural disasters) and human activities [1]. Some of the most common and harmful types of marine debris include plastics ranging greatly in size, such as bags, bottles, fishing gear, and microplastic particles.

**Macroplastics**

Within the marine context, millions of tonnes of plastic enter our oceans annually as micro- to macroplastic litter, as a result, plastic now accounts for more than 80–90% of all litter found in this environment.

Macroplastics describe large plastic items with a diameter ≥ 5 mm. In many instances, these items can persist for decades in the ocean especially when shielded from UV radiation under water or in sediments, accounting for a large portion of the global mass of floating debris [2], [3].

**Sargassum**

*Sargassum* is a genus of large brown seaweed (a type of algae) that is abundant in the ocean. Floating rafts of this seaweed can stretch for several kilometers across the ocean providing food and refuge for several animals, however when it reaches the coast it can affect tourism and consequently the local economy.

Although *sargassum* is not included in the definition of marine debris, it is important during the detection and classification processes to separate debris from floating vegetation.

### Detection and Classification Methods

A satellite-based marine debris detection system needs to screen out all non-water pixels (land and cloud masks may be used), locate pixels containing floating material and successfully discriminate between anthropogenic debris (e.g. plastics) and natural phenomena (e.g. floating vegetation). This can be achieved using a variety of techniques, from band indices (e.g. FAI and FDI) to linear unmixing using spectral signatures, and machine learning. Satellite remote sensors are the primary observation tool and need to be selected in terms of revisit time, spatial resolution, spectral availability and radiometric quality [3].

When floating debris and vegetation are aggregated into sufficiently large patches in the marine environment,
detection from Sentinel-2 is possible. The 13-band multispectral instrument (MSI) optical sensors aboard the Sentinel-2 satellites provide revisit times of 2 to 5 days and spatial resolution of up to 10m allowing macroplastics to be distinguishable from natural sources of debris, and seawater itself. This distinction is also possible due to the spectral properties of the materials as we can see in Fig.4 [2].

In this workshop we will consider the detection technique through band indices.

**Floating Algae Index (FAI)**

Water strongly absorbs light in the red, near infrared (NIR) and shortwave infrared (SWIR) wavelengths, while vegetation floating on the water surface have higher reflectance in the NIR than in other wavelengths. These characteristics allow to easily distinguish algae from surrounding clear water. Equations 1 and 2 translate this and are used to calculate FAI, an index considered to be an improved version of the NDVI (Normalized Difference Vegetation Index) [4].

**Floating Debris Index (FDI)**

At the highest spatial resolution of the Sentinel-2 MSI, individual items of debris are likely to be below detectable limits until aggregated into large patches. To enhance detection of patches floating on the ocean in Sentinel-2 imagery, the FDI was developed using four MSI bands. FDI is a modified version of the FAI where, in place of the red band, where chlorophyll is most absorptive, the Red Edge 2 band is used (see equations 3 and 4). The FDI thus leverages the difference between the NIR, and the baseline reflectance of NIR, presenting good potential in offshore plastic detection [2].

### About this Jupyter Notebook

This notebook shows how you can automatically download Sentinel-2 products using Python and process them to identify agglomerations of floating vegetation (such as *sargassum*) and possible clusters of marine litter.

This workflow is a simplified version of what is being developed for the *LABPLAS* (Land-Based Solutions for Plastics in the Sea) project, being a first step in the identification of marine debris.  The classification into macroplastics, floating vegetation, timber, pumice rock, foam, etc... must be done in a later stage using a spectral signatures library and machine learning technics.

### Notes

**1 -** We will be using Sentinelsat (a library for Python) to search and download Sentinel-2 products from the [Copernicus Open Access Hub](https://scihub.copernicus.eu/dhus/#/home) (COAH). However, sometimes these products are offline and their recovery must be done from the LTA (Long Term Archive). This process can take some time and is not always successful, so you can use DIASs (Data and Information Access Services) to check the availability of a product and download it manually or automatically. Some DIASs examples include:

* [CREODIAS](https://creodias.eu)
* [ONDA DIAS](https://www.onda-dias.eu/cms/)

**2-** You will need to create a free account on [Copernicus Open Access Hub](https://scihub.copernicus.eu/dhus/#/self-registration) if you want to use the Sentinelsat library.

**3-** The entire workflow is adapted to the projection used by the Sentinel-2 product (UTM). Your shapefile will be converted automatically.

The Universal Transverse Mercator system divides the Earth's surface into 60 zones. Each UTM zone has a vertical width of 6° of longitude and horizontal width of 8° of latitude.

**4-** This script only works with one S2L2A product at a time. You can adapt the code to download multiple products and merge them all together using GDAL.

**5-** The handling of some errors is ignored. A good idea, if you want to run this code in a server, is to send the error alert by e-mail using smtplib and email-to.
